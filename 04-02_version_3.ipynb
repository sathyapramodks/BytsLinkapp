```
df_temp = df_combined[['17_x', '18_x', '19_x', '17_y', '18_y', '19_y']].fillna(0)
```

```
for i in df_temp.columns:
  df_combined[i] = df_temp[i]
```

```
df_combined.drop(columns={'warehouse_stock'}, inplace = True)
```

```
def sub(x):
  return {'Variance_17': int(x['17_x'] - x['17_y'])
          ,'Variance_18': int(x['18_x'] - x['18_y'])
          ,'Variance_19': int(x['19_x'] - x['19_y'])
  }
```

```
def sub_2(x, lst):
  return int(x[lst[0]] - x[lst[1]])
```

```
df_combined['variance_17'] = df_combined.apply(sub_2, axis = 1, lst = ['17_x', '17_y'])
df_combined['variance_18'] = df_combined.apply(sub_2, axis = 1, lst = ['18_x', '18_y'])
df_combined['variance_19'] = df_combined.apply(sub_2, axis = 1, lst = ['19_x', '19_y'])
```

```
df_combined.to_csv('df_combined.csv')
files.download('df_combined.csv')
```
